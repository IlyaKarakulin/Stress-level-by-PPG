```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt
import os 
import pickle
```

```python
data = pd.read_pickle(f'/home/ilya/Documents/stress-level-by-PPG/data.pkl')
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(200, 6))
ax1.plot(data['S2']['PPG_stress'], color='red')
ax1.grid(True)
ax1.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
ax2.plot(data['S2']['ACC_stress'])
ax2.grid(True)  
ax2.set_ylim(-128, 127)
ax2.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(200, 6))
ax1.plot(data['S2']['PPG_medit'], color='red')
ax1.grid(True)
ax1.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
ax2.plot(data['S2']['ACC_medit'])
ax2.grid(True)
ax2.set_ylim(-128, 127)
ax2.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

plt.show()
```
